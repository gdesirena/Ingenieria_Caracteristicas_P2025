## 2.- Medidas de posición y de Forma

Considere el dataset tips que se encuentra en la libreria sns. Realice lo que se pide.

1. Cargar el dataset (sns.load_dataset('tips')) y mostrar las primeras filas, obtenga la asimetría y la curtósis de cada una de las características de este dataset.
2. Obtengha una visualización de la distribución y medidas de forma de la caracterísitca total_bill
3. Compara la asimetría y curtósis de total_bill por día de la semana
4. Obtenga los cuartiles de la columna "tip".
5. Obtenga un diagrama de cajas para la variable tip, ¿qué puede comentar sobre este gráfico?


```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Sub-ejercicio 2.1: Cargar el dataset tips y mostrar las primeras filas
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Sub-ejercicio 2.2: Medidas de forma: asimetría y curtosis de total_bill
from scipy.stats import skew, kurtosis
total_bill = tips['total_bill']
asimetria = skew(total_bill)
curtosis_valor = kurtosis(total_bill, fisher=False)
print(f'Asimetría: {asimetria:.2f}\nCurtosis: {curtosis_valor:.2f}')
```

```python
# Sub-ejercicio 2.3: Medidas de forma para la variable tip
tip = tips['tip']
asimetria_tip = skew(tip)
curtosis_tip = kurtosis(tip, fisher=False)
print(f'Asimetría (tip): {asimetria_tip:.2f}\nCurtosis (tip): {curtosis_tip:.2f}')
```

```python
# Sub-ejercicio 2.4: Visualización de la distribución y medidas de forma de total_bill
plt.figure(figsize=(8,5))
plt.hist(total_bill, bins=20, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribución de total_bill (tips)')
plt.xlabel('Total de la cuenta ($)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Sub-ejercicio 2.5: Comparar la asimetría y curtosis de total_bill por día de la semana
dias = tips.groupby('day')['total_bill'].agg(['mean', 'std', skew, kurtosis])
dias.rename(columns={'skew': 'asimetria', 'kurtosis': 'curtosis'}, inplace=True)
dias
```

```python
# Medidas de posición: cuartiles y percentiles de tip
q1 = np.percentile(tip, 25)
q2 = np.percentile(tip, 50)
q3 = np.percentile(tip, 75)
p90 = np.percentile(tip, 90)
print(f'Q1 (25%): {q1:.2f}\nQ2 (Mediana, 50%): {q2:.2f}\nQ3 (75%): {q3:.2f}\nPercentil 90: {p90:.2f}')
```

## 3.- Correlación y Relaciones lineales

Considere el dataset iris que se encuentra en la libreria sns. Realice lo que se pide.

1. Cargar el dataset (sns.load_dataset('iris')) y mostrar las primeras filas.
2. Calcule la correlación de las características del dataset.
3. Obtenga un modelo matemático que relacione por pares, las características del dataset.
4. Según lo encontrado, ¿qué puede decir sobre lass relaciones que existen en el conjunto de datos?.

```python
#Cargar el dataset (sns.load_dataset('iris')) y mostrar las primeras filas.
import seaborn as sns
import pandas as pd
iris = sns.load_dataset('iris')
iris.head()
```

```python
#Calcule la correlación de las características del dataset.
correlaciones = iris.corr(numeric_only=True)
correlaciones
```

```python
# Visualización de la matriz de correlación
import matplotlib.pyplot as plt
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de correlación del dataset Iris')
plt.show()
```

```python
#Según lo encontrado, ¿qué puede decir sobre lass relaciones que existen en el conjunto de datos?.
from itertools import combinations
from sklearn.linear_model import LinearRegression
import numpy as np

caracteristicas = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
resultados = []

for x_col, y_col in combinations(caracteristicas, 2):
    X = iris[[x_col]].values
    y = iris[y_col].values
    modelo = LinearRegression().fit(X, y)
    pendiente = modelo.coef_[0]
    interseccion = modelo.intercept_
    score = modelo.score(X, y)
    resultados.append({
        'X': x_col,
        'Y': y_col,
        'pendiente': pendiente,
        'interseccion': interseccion,
        'R2': score
    })

pd.DataFrame(resultados)
```

- Las correlaciones más altas se observan entre `petal_length` y `petal_width`, lo que indica una fuerte relación lineal positiva entre estas variables.
- `sepal_length` también tiene una correlación positiva considerable con `petal_length` y `petal_width`.
- `sepal_width` muestra correlaciones más bajas e incluso negativas con otras características.
- Los modelos de regresión lineal muestran que es posible predecir una característica a partir de otra con cierto grado de precisión, especialmente entre las variables de pétalos.

**Conclusión:**
El dataset Iris presenta relaciones lineales fuertes entre algunas de sus características, especialmente entre las dimensiones de los pétalos. Esto sugiere que estas variables están relacionadas y pueden ser útiles para tareas de clasificación y análisis predictivo.
