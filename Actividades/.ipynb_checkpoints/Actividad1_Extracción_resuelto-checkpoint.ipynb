## 1.- JSON

Supón que tienes un archivo `empleados.json` con la siguiente estructura:

```json
[
  {"nombre": "Ana", "edad": 28, "departamento": "Ventas", "salario": 2500},
  {"nombre": "Luis", "edad": 35, "departamento": "IT", "salario": 3200},
  {"nombre": "María", "edad": 24, "departamento": "Ventas", "salario": 2300},
  {"nombre": "Carlos", "edad": 40, "departamento": "IT", "salario": 4000}
]
```

1. Crea un dataframe apartir de la extracción de información de este archivo json
2. ¿Cuál es el salario promedio por departamento?
3. ¿Quién es el empleado más joven y a qué departamento pertenece?
4. ¿Cuántos empleados tienen un salario mayor a 2500?
5. ¿Cuál es la diferencia de edad entre el empleado más joven y el de mayor edad?
6. ¿Cuál es el salario total pagado en la empresa?


````python
import pandas as pd
import json

# Simulación de lectura de archivo JSON
json_data = '[{"nombre": "Ana", "edad": 28, "departamento": "Ventas", "salario": 2500}, {"nombre": "Luis", "edad": 35, "departamento": "IT", "salario": 3200}, {"nombre": "María", "edad": 24, "departamento": "Ventas", "salario": 2300}, {"nombre": "Carlos", "edad": 40, "departamento": "IT", "salario": 4000}]'
empleados = json.loads(json_data)
df_json = pd.DataFrame(empleados)
print(df_json)
````

````python
print('Salario promedio por departamento:')
print(df_json.groupby('departamento')['salario'].mean())
joven = df_json.loc[df_json['edad'].idxmin()]
print(f'Empleado más joven: {joven["nombre"]}, Departamento: {joven["departamento"]}')
print('Empleados con salario > 2500:', (df_json['salario'] > 2500).sum())
print('Diferencia de edad:', df_json['edad'].max() - df_json['edad'].min())
print('Salario total pagado:', df_json['salario'].sum())
````

## 2.- XML

Supón que tienes un archivo `libros.xml` con la siguiente estructura:

```xml
<libros>
  <libro>
    <titulo>Python para Todos</titulo>
    <autor>Juan Pérez</autor>
    <precio>29.99</precio>
    <anio>2020</anio>
  </libro>
  <libro>
    <titulo>Data Science Avanzado</titulo>
    <autor>María López</autor>
    <precio>45.50</precio>
    <anio>2022</anio>
  </libro>
  <libro>
    <titulo>Ingeniería de Características</titulo>
    <autor>Pedro Ruiz</autor>
    <precio>39.00</precio>
    <anio>2019</anio>
  </libro>
</libros>
```
1. Crea un dataframe apartir de la extracción de información de este archivo.
2. ¿Cuál es el libro más caro y su autor?
3. ¿Cuál es el precio promedio de los libros publicados después de 2020?
4. ¿Cuántos libros fueron publicados antes de 2021?
5. ¿Cuál es la diferencia de precio entre el libro más barato y el más caro?
6. ¿Cuántos autores diferentes hay en la lista?

````python
import xml.etree.ElementTree as ET

xml_data = '''
<libros>
  <libro>
    <titulo>Python para Todos</titulo>
    <autor>Juan Pérez</autor>
    <precio>29.99</precio>
    <anio>2020</anio>
  </libro>
  <libro>
    <titulo>Data Science Avanzado</titulo>
    <autor>María López</autor>
    <precio>45.50</precio>
    <anio>2022</anio>
  </libro>
  <libro>
    <titulo>Ingeniería de Características</titulo>
    <autor>Pedro Ruiz</autor>
    <precio>39.00</precio>
    <anio>2019</anio>
  </libro>
</libros>
'''
root = ET.fromstring(xml_data)
libros = []
for libro in root.findall('libro'):
    titulo = libro.find('titulo').text
    autor = libro.find('autor').text
    precio = float(libro.find('precio').text)
    anio = int(libro.find('anio').text)
    libros.append({'titulo': titulo, 'autor': autor, 'precio': precio, 'anio': anio})
df_xml = pd.DataFrame(libros)
print(df_xml)
````

````python
caro = df_xml.loc[df_xml['precio'].idxmax()]
print(f'Libro más caro: {caro["titulo"]}, Autor: {caro["autor"]}')
print('Precio promedio de libros publicados después de 2020:', df_xml[df_xml['anio'] > 2020]['precio'].mean())
print('Libros publicados antes de 2021:', (df_xml['anio'] < 2021).sum())
print('Diferencia de precio:', df_xml['precio'].max() - df_xml['precio'].min())
print('Número de autores diferentes:', df_xml['autor'].nunique())
````

## 3.- CSV

Considera el archivo `ventas_detalladas.csv` con la siguiente estructura:

```csv
producto,fecha,cantidad,precio_unitario
A,2023-01-01,10,2.5
B,2023-01-02,5,3.0
A,2023-01-03,7,2.5
C,2023-01-01,3,4.0
B,2023-01-03,2,3.0
```

````python
import io
csv_data = '''producto,fecha,cantidad,precio_unitario\nA,2023-01-01,10,2.5\nB,2023-01-02,5,3.0\nA,2023-01-03,7,2.5\nC,2023-01-01,3,4.0\nB,2023-01-03,2,3.0'''
df_csv = pd.read_csv(io.StringIO(csv_data), parse_dates=['fecha'])
df_csv.to_csv('ventas_detalladas.csv')
print(df_csv)
````

### Preguntas de análisis CSV
1. Crea un dataframe con los datos extraidos del archivo en cuestión.
2. ¿Cuál es el total vendido por producto y por fecha?
3. ¿En qué fecha se vendió la mayor cantidad de productos?
4. ¿Cuál es el producto con mayor ingreso total?
5. ¿Cuál es la media de cantidad vendida por producto?
6. ¿Qué producto tuvo ventas todos los días registrados?

````python
df_csv['total'] = df_csv['cantidad'] * df_csv['precio_unitario']
print('Total vendido por producto y fecha:')
print(df_csv.groupby(['producto', 'fecha'])['total'].sum())
fecha_max = df_csv.groupby('fecha')['cantidad'].sum().idxmax()
print('Fecha con mayor cantidad vendida:', fecha_max)
ingreso_total = df_csv.groupby('producto')['total'].sum()
print('Producto con mayor ingreso total:', ingreso_total.idxmax())
print('Media de cantidad vendida por producto:')
print(df_csv.groupby('producto')['cantidad'].mean())
ventas_por_dia = df_csv.groupby('producto')['fecha'].nunique()
dias_registrados = df_csv['fecha'].nunique()
productos_todos_dias = ventas_por_dia[ventas_por_dia == dias_registrados].index.tolist()
print('Productos con ventas todos los días:', productos_todos_dias)
````

## 4.- Imágenes

Considera la imagen `iteso.jpg` que se encuentra en la misma ruta que este archivo de jupyter.

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Crear una imagen RGB artificial para el ejemplo
imagen_array = np.zeros((100, 100, 3), dtype=np.uint8)
imagen_array[20:80, 20:80] = [0, 0, 255]  # Cuadro azul en el centro
imagen = Image.fromarray(imagen_array)

# Mostrar la imagen
plt.imshow(imagen)
plt.axis('off')
plt.title('Imagen RGB de ejemplo')
plt.show()
````

1. Extrae la información de los pixeles de la imagen.
2. ¿Cuál es el valor RGB del píxel (50, 50)?
3. ¿Cuántos píxeles tienen el color azul puro (0, 0, 255)?
4. ¿Cuál es el color predominante en la imagen?


````python
imagen_np = np.array(imagen)
print('Valor RGB del píxel (50, 50):', imagen_np[50, 50])
azul_puro = np.sum(np.all(imagen_np == [0, 0, 255], axis=-1))
print('Cantidad de píxeles azul puro:', azul_puro)
valores, cuentas = np.unique(imagen_np.reshape(-1, 3), axis=0, return_counts=True)
color_predominante = valores[cuentas.argmax()]
print('Color predominante (RGB):', color_predominante)
negros = np.sum(np.all(imagen_np == [0, 0, 0], axis=-1))
print('Proporción de píxeles negros:', negros / imagen_np.size * 3)
print('Área del cuadro azul:', azul_puro)
````
